```python
""" 
Verificando consistência entre as colunas do dataset
 """
dfs = [
    df19_20,
    df20_21,
    df21_22,
    df22_23,
    df23_24,
    df24_25
]

for i in range(len(dfs)-1):
    df1 = dfs[i]
    for j in range(i+1, len(dfs)):
        df2 = dfs[j]

        print(f"\n\n{df1['Date'][0].split(sep='/')[-1]} vs {df2['Date'][0].split(sep='/')[-1]}")

        # Colunas diferentes entre 2 datasets
        for col in df1.columns:
            if col not in df2.columns:
                print(f"\t{col}", sep= " ")
        
```

O dataset de 24/25 além de ter colunas extras, também não apresenta algumas colunas dos datasets antigos. Todas essas inconsistências não são relevantes de manter para o trabalho. 
- Representam casas de apostas extras. (Ver [legenda](legenda_colunas.md) para consultar as colunas a serem consideradas.)

```python
for df in dfs:
    print(f" {df['Date'].iloc[0]} {"\n"} {df.info(verbose= True, show_counts=True)} {"\n\n\n"}")
```

```python
df_full = pd.concat(dfs, ignore_index= True)

df_full.iloc[[0,-1]]
```

```python
# Checando consistência da escolha

df_full[selected_features].info(verbose= True, show_counts=True)
```

Praticamente todas as colunas estão completas. "P>2.5 e P<2.5", odds da Pinnacle para mais de 2.5 gols possuem 10 instâncias não preenchidas que podem ser dropadas.
